```python
import pandas as pd
from pandas import DataFrame
df = pd.read_csv('D:\Data Mining\Lab\Lab6\store_data.csv',header=None)
```

```python
records = []  
for i in range(0, 7501):  
    records.append([str(df.values[i,j]) for j in range(0, 20)])
records
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(records).transform(records)
df1 = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets=apriori(df1, min_support=0.05, use_colnames=True)
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
```
