```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./Datasets/tips.csv')
```

```python
data.head()
```

```python
data.info()
```

## Dist plot

```python
sns.set_style('whitegrid')
sns.distplot(data['total_bill'], kde = True, color = 'b', bins = 30)
```

## Joint plot

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = data, kind = 'reg', color = 'b')
```

```python
sns.pairplot(data, hue = 'sex', palette = 'rainbow_r', markers = '+', kind = "reg")
```

## Rug plot

```python
sns.kdeplot(data['total_bill'], data['tip'])
sns.rugplot(data['total_bill'])
```
